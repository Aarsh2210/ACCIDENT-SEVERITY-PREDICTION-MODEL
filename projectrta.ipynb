```python
data['Time'] = pd.to_datetime(data['Time'])


new_data = data.copy()
new_data['Hour_of_Day'] = new_data['Time'].dt.hour
n_data = new_data.drop('Time', axis=1)
n_data.head()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(n_data.isnull(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Heatmap of Missing Values")
plt.show()
```

```python
#Dealing with Missing values
missing_percentage = n_data.isnull().mean() * 100
print("\nMissing percentage in each column:")
print(missing_percentage)

n_data = n_data.loc[:, missing_percentage <= 50]
    
   
numerical_cols = n_data.select_dtypes(include=['float64', 'int64']).columns
n_data[numerical_cols] = n_data[numerical_cols].fillna(n_data[numerical_cols].mean())

    
categorical_cols = n_data.select_dtypes(include=['object']).columns
n_data[categorical_cols] = n_data[categorical_cols].fillna(n_data[categorical_cols].mode().iloc[0])

```

```python
sns.heatmap(n_data.isnull(), cmap="viridis", cbar=False, yticklabels=False)
plt.title("Missing Values After Imputation")
plt.show()
```

```python
#Feature Engineering
features = ['Day_of_week','Number_of_vehicles_involved','Number_of_casualties','Area_accident_occured',
           'Types_of_Junction','Age_band_of_driver','Sex_of_driver','Educational_level',
           'Vehicle_driver_relation','Type_of_vehicle','Driving_experience','Service_year_of_vehicle','Type_of_collision',
           'Sex_of_casualty','Age_band_of_casualty','Cause_of_accident','Hour_of_Day']


X = n_data[features]  
y = n_data['Accident_severity']


encoded_data = pd.get_dummies(X, drop_first=True)
encoded_data.shape

```

```python
target_encoded=pd.get_dummies(y, drop_first=True)
target_encoded.shape
    
```

```python


from sklearn.feature_selection import SelectKBest, chi2
fs = SelectKBest(chi2, k=50)
X_new = fs.fit_transform(encoded_data, target_encoded)

cols = fs.get_feature_names_out()


fs_df = pd.DataFrame(X_new, columns=cols)



```

```python

#Using Smote for oversampling
from imblearn.over_sampling import SMOTE



target_labels = np.argmax(target_encoded.values, axis=1)  


smote = SMOTE(random_state=42)


X_n, y_n = smote.fit_resample(fs_df, target_labels)


print(X_n.shape, y_n.shape)




```

```python
y_n_series = pd.Series(y_n)


class_counts = y_n_series.value_counts()

print(class_counts)
```

```python
#Using Random Forest Model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report,f1_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_n, y_n, test_size=0.2, random_state=42)
```

```python
rf = RandomForestClassifier(n_estimators=800, max_depth=20, random_state=42)
rf.fit(X_train, y_train)


predicted = rf.predict(X_test)


rf.score(X_train, y_train)

```

```python

classif_report = classification_report(y_test,predicted)
print(classif_report)
```

```python
f1score = f1_score(y_test,predicted, average='weighted')
print(f1score)
```

```python
#Using KNeighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier
Kf=KNeighborsClassifier(n_neighbors=8)
Kf.fit(X_train,y_train)
predicted = Kf.predict(X_test)


Kf.score(X_train, y_train)



```

```python
classif_report = classification_report(y_test,predicted)
print(classif_report)
```

```python
f1score = f1_score(y_test,predicted, average='weighted')
print(f1score)
```

```python
#Using SVM model
from sklearn.svm import SVC


svm = SVC(kernel='rbf', random_state=40) 
svm.fit(X_train, y_train)
predicted = svm.predict(X_test)


svm.score(X_train, y_train)
```

```python
classif_report = classification_report(y_test,predicted)
print(classif_report)
```

```python
f1score = f1_score(y_test,predicted, average='weighted')
print(f1score)
```
